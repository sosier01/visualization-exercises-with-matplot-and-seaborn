```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris_df = sns.load_dataset('iris')
```

```python
iris_df.head()
```

```python
# Which flowers are biggest on average? what does biggest mean? which flowers have the biggest size variation?
# for petals and sepals, length and width
# is there any correlation b/w sepal and petal size?
```

```python
# The biggest flowers would be the flower with the largest petal surface area, adding a column for that
petal_size = iris_df['petal_length'] * iris_df['petal_width']
iris_df['petal_size'] = petal_size
iris_df.head()
```

```python
average_petal_size = iris_df.groupby('species')['petal_size'].mean()
average_petal_size
```

```python
sepal_size = iris_df['sepal_length'] * iris_df['sepal_width']
iris_df['sepal_size'] = sepal_size
average_sepal_size = iris_df.groupby('species')['sepal_size'].mean()
```

```python
sns.barplot(average_petal_size)
```

```python
sns.barplot(average_sepal_size)
```

```python
sns.scatterplot(x='petal_size', y='sepal_size', data = iris_df, hue = 'species', palette= '2')
```

```python
sns.pairplot(iris_df, hue="species")
plt.show()
```
